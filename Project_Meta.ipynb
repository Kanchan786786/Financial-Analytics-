```python
import yfinance as yf
import pandas as pd

# Define the stock symbol and date range
meta_ticker = 'META'
start_date = '2022-01-01'
end_date = '2024-12-31'

# Fetch stock data
meta_stock = yf.Ticker(meta_ticker)

# Get historical stock price data (for technical analysis, Monte Carlo, and forecasting)
historical_prices = meta_stock.history(period='1d', start=start_date, end=end_date)

# Get dividend history (for valuation models and dividend growth)
dividends = meta_stock.dividends.loc[start_date:end_date]

# Get stock splits (useful for adjusting historical prices)
splits = meta_stock.splits.loc[start_date:end_date]

# Get balance sheet (for ratio calculations)
balance_sheet = meta_stock.balance_sheet

# Get income statement (for profitability ratios and sustainable growth rate)
income_statement = meta_stock.financials

# Get cash flow statement (for liquidity ratios)
cash_flow = meta_stock.cashflow

# Save data to CSV files
historical_prices.to_csv('meta_historical_prices.csv')
dividends.to_csv('meta_dividends.csv')
splits.to_csv('meta_splits.csv')
balance_sheet.to_csv('meta_balance_sheet.csv')
income_statement.to_csv('meta_income_statement.csv')
cash_flow.to_csv('meta_cash_flow.csv')

# Preview the data
print("Historical Prices (Head):")
print(historical_prices.head())

print("\nDividends:")
print(dividends)

print("\nStock Splits:")
print(splits)

print("\nBalance Sheet:")
print(balance_sheet.head())

print("\nIncome Statement:")
print(income_statement.head())

print("\nCash Flow Statement:")
print(cash_flow.head())

# Combine into a single summary DataFrame
meta_data_summary = pd.DataFrame({
    'Date': historical_prices.index,
    'Close Price': historical_prices['Close'],
    'Volume': historical_prices['Volume'],
    'Dividend': dividends.reindex(historical_prices.index, fill_value=0),
    'Stock Split': splits.reindex(historical_prices.index, fill_value=1)
})

# Save combined summary
meta_data_summary.to_csv('meta_stock_summary.csv')

print("\nCombined Stock Data Summary:")
print(meta_data_summary.head())

# Notes:
# 1. Historical Prices: Suitable for technical analysis, Monte Carlo simulations, and forecasting (Prophet)
# 2. Dividends: Essential for valuation models (CAPM, dividend growth model)
# 3. Splits: Useful for adjusting historical prices and calculating accurate returns
# 4. Balance Sheet: Needed for ratio calculations (liquidity, debt-to-equity, sustainable growth rate)
# 5. Income Statement: Useful for profitability ratios and growth metrics
# 6. Cash Flow Statement: Needed for liquidity ratios and financial health assessment

# Let me know if youâ€™d like me to refine or add more datasets for specific parts of your project! ðŸš€
```
